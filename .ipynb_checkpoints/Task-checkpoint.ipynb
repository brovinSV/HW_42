```python
"""
Застосуйте описані вище методи кластеризації на датасеті винних сортів UCI ML Wine Data Set. 
Завантажити його можна командою datasets.load_wine. Результати кластеризацій оформіть у вигляді файлу ipynb.
"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# завантаження даних
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
wine_data = pd.read_csv(url, header=None)
wine_data.columns = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium',
                     'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                     'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
```

```python
# Видалення стовпця з класом
X = wine_data.drop('Class', axis=1)
```

```python
# Стандартизація даних
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# Зниження розмірності до 2D для візуалізації
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape
```

```python
# Кластеризація методом k-середніх
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
kmeans.cluster_centers_
wine_centers = pca.transform(kmeans.cluster_centers_)
kmeans_labels
```

```python
# Кластеризація методом DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)

dbscan_labels = dbscan.fit_predict(X_scaled)
dbscan_labels
```

```python
# Кластеризація ієрархічним методом
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(X_scaled)
hierarchical_labels
```

```python
# Візуалізація результатів
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels)
dots = plt.scatter(wine_centers[:,0], wine_centers[:,1], s=100, c='r')
plt.title("K-Means Clustering")

plt.subplot(132)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=dbscan_labels)
plt.title("DBSCAN Clustering")

plt.subplot(133)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=hierarchical_labels)
plt.title("Hierarchical Clustering")

plt.tight_layout()
plt.show()
```
